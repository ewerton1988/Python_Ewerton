###**AULA 2 - Biblioteca Matplotlib: para visualização dos dados**

```python
#importando a biblioteca numpy
import numpy as np

#Importanto a biblioteca: no pacote pyplot se encontra todos os recursos de matplotlib
from matplotlib import pyplot as plt

#utilizado em ambiente notebook para que os gráficos sejam renderizados
%matplotlib inline 
```

###**Função scatter**

Nossa plotagem do nosso primeiro gráfico usando a biblioteca de funções matplotlib.
Para se fazer um gráfico, precisamos de pelo menos coordenadas em nosso plano cartesiano. Para isso, devemos plotá-los individualmente:

```python
x = 5 #eixo x
y = 2 #eixo y

#plotando o ponto usando a função scatter
plt.scatter(x,y)
```

**Plotando uma lista de pontos**

```python
x = [2, 4, 6, 8, 10] #eixo x
y = [1, 3, 5, 7, 9] #eixo y

#plotando as listas de coordenadas para cada eixo usando a função scatter
plt.scatter(x,y)
```

###**Função plot**

Vamos agora utilizar a mesma lista para plotar o nosso gráfico de outra forma. A função scatter plota unicamente todos os pontos informados a ela, mas há outra função que plota um gráfico que liga todos os pontos, na respectiva ordem informada, vamos utilizar a função plot.

```python
x = np.array([2, 4, 6, 8, 10]) #eixo x
y = np.array([1, 3, 5, 7, 9]) #eixo y

#plotando as listas de coordenadas para cada eixo usando a função plot
plt.plot(x,y)
plt.show()
```

###**Personalizando o gráfico: legenda, título, cor, grade**

```python
plt.plot(x, y)

#Definindo título
plt.title('Plotando uma reta com matplotlib')

#Definindo labels
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')

plt.show()
```

```python
#alterando a cor da reta para vermelho
plt.plot(x, y, color = 'red')

#Definindo título
plt.title('Plotando uma reta vermelha com matplotlib')

#Definindo labels
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')

plt.show()
```

```python
#alterando a cor da reta para vermelho
plt.plot(x, y, color = 'red')

#Definindo título
plt.title('Plotando uma reta vermelha com matplotlib e mostrando grade')

#Definindo labels
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')

#Mostrando grade no gráfico
plt.grid(True)

plt.show()
```

###**Altura e largura do gráfico**

```python
#definindo tamanho do gráfico
plt.figure(figsize=(10,5)) #(altura,largura)

plt.plot(x, y)

#Definindo título
plt.title('Plotando uma reta com matplotlib')

#Definindo labels
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')

plt.show()
```

```python
#Plotagem 1D
x = np.linspace(0, 3, 20)
y = np.linspace(0, 9, 20)

plt.plot(x, y)       # plotagem da linha
plt.plot(x, y, 'o')  # plotagem dos pontos
plt.show()           # <-- mostra o gráfico
```

###**Plotando dados categóricos**

Trabalhando com gráficos de barras.
Ideal para quando estamos trabalhando com comparação de valores.
Normalmente quer se visualizar frequência ou percentual de valores.

**Plotando gráfico de barras verticais o gráfico de colunas**

```python
#Para plotar um gráfico de barras usamos o método bar que recebe 2 parâmetros, 
#a lista com as categorias dos produtos (x) e uma lista com os seus respectivos valores (y).

#Cidades com maior número de população em milhar
cidades = ['Santo André', 'São Bernardo', 'Mauá']
populacao = [700, 1000, 500]

#plt.figure(figsize=(8,8)) #(altura,largura)
plt.bar(cidades, populacao, color='Red', label= 'População em milhar')
plt.title('Cidades mais populosas do ABC')

plt.legend()
plt.show()
```

**Plotando Gráfico de barras horizontais**

Gráfico muito usado quando queremos fazer comparações.
Pode ser usado para expressar uma ideia cronológica.
Leitura muito simples e intuitiva.
Funciona bem quando os labels dos eixos são longos.

```python
#Plotando o gráfico de barras na horizontal plt.barh

plt.barh(cidades, populacao, color='g', label= 'População em milhar')
plt.title('Cidades mais populosas do ABC')

plt.legend()
plt.show()
```

###**Gráfico de Pizza**

Gráfico muito utilizado para mostrar a proporção de sexo (feminino / masculino) e para faixa etária (criança, jovem, adulto, idoso).

Mas se você tiver muitos dados (muitas categorias por exemplo) evite usar esse tipo de gráfico, pois a visualização fica muito confusa.

```python
#Cidades com maior número de população em milhar
cidades = ['Santo André', 'São Bernardo', 'Mauá']
populacao = [700, 1000, 500]

plt.pie(populacao, labels=cidades)
plt.show()
```

**Customizando gráfico de pizza: Adicionando informações de percentual, sombra, separação das partes e legenda.**

```python
# define o nível de separabilidade entre as partes, ordem do vetor representa as partes
explode = (0, 0.1, 0) 

# define o formato de visualização com saída em 1.1%%, sombras e a separação entre as partes
plt.pie(populacao, labels=cidades, autopct='%1.1f%%', shadow=True, explode=explode)


# inseri a legenda e a localização da legenda.
plt.legend(cidades, loc=9)

plt.show()
```

Colocamos a quantidade em valores percentuais dando destaque (explode=explode) para a categoria que teve o maior valor. O método shadow=True é responsável pelo sombreamento dando um aspecto de 3D.

A localização da legenda é definida dentro de um quadrante:



*   loc=1 → canto superior direito
*   loc=2 → canto superior esquerdo
*   loc=3 → canto inferior esquerdo
*   loc=4 → canto inferior direito

###**Plotando Histogramas**

Histogramas são úteis quando queremos visualizar a distribuição de uma variável.
Esse é um gráfico muito utilizado em estatítica descritiva.

É útil em tarefas de análise e exploração de dados.

```python
plt.hist(populacao, bins=10,color='b')
```
